```python
from pyace import create_multispecies_basis_config
from pyace.multispecies_basisextension import single_to_multispecies_converter, create_multispecies_basis_config
from pyace.generalfit import save_interim_potential
```

### Making empty potential for H

#### empty_H_pot

```python
potential_config={
    'deltaSplineBins': 0.001,
    'elements': ['H'],
    'embeddings': {'ALL': {'drho_core_cut': 500,
                           'fs_parameters': [1, 1, 1, 0.5],
                           'ndensity': 2,
                           'npot': 'FinnisSinclair',
                           'rho_core_cut': 3000},
                   },

    'bonds': {'ALL': {'NameOfCutoffFunction': 'cos',
                      'dcut': 0.02,
                      'radbase': 'SBessel',
                      'radparameters': [5.25],
                      'r_in': 0.3,
                      'delta_in': 0.1,
                      'core-repulsion': [1000, 0],
                      'rcut': 6.4},
              },

    'functions': {   
        'number_of_functions_per_element': 800,        
        'H': {
            'nradmax_by_orders': [15, 10, 6, 3],
            'lmax_by_orders':    [0,  8, 4, 2]
        },
        

        },
    }
```

```python
# H_bbasis=create_multispecies_basis_config(potential_config = single_to_multispecies_converter(potential_config))
H_bbasis=create_multispecies_basis_config(potential_config)
```

```python
H_bbasis
```

```python
H_bbasis.total_number_of_functions
```

```python
save_interim_potential(H_bbasis, potential_filename='./empty_H.yaml')
```

#### empty_HSi_pot

```python
potential_config={
    'deltaSplineBins': 0.001,
    'elements': ['H','Si'],
    'embeddings': {'ALL': {'drho_core_cut': 500,
                           'fs_parameters': [1, 1, 1, 0.5],
                           'ndensity': 2,
                           'npot': 'FinnisSinclair',
                           'rho_core_cut': 3000},
                   },

    'bonds': {'ALL': {'NameOfCutoffFunction': 'cos',
                      'dcut': 0.02,
                      'radbase': 'SBessel',
                      'radparameters': [5.25],
                      'r_in': 0.35,
                      'delta_in': 0.1,
                      'core-repulsion': [10000, 0],
                      'rcut': 6.4},

              },
    "functions": {
        "number_of_functions_per_element": 1200,
        "UNARY": {
            "nradmax_by_orders": [20,12,8,6,3,1,0],
            "lmax_by_orders": [0,7,6,4,2,1,0],
            "coeffs_init": ""
        },
        "BINARY": {
            "nradmax_by_orders": [15,10,7,5,3],
            "lmax_by_orders": [0,8,6,4,2],
            "coeffs_init": ""
        },

    }
}
```

```python
HSi_bbasis=create_multispecies_basis_config(potential_config)
```

```python
HSi_bbasis
```

```python
HSi_bbasis.total_number_of_functions
```

```python
save_interim_potential(HSi_bbasis, potential_filename='./empty_HSi.yaml')
```

#### empty_Si_pot

```python
potential_config={
    'deltaSplineBins': 0.001,
    'elements': ['Si'],
    'embeddings': {'ALL': {'drho_core_cut': 500,
                           'fs_parameters': [1, 1, 1, 0.5],
                           'ndensity': 2,
                           'npot': 'FinnisSinclair',
                           'rho_core_cut': 3000},
                   },

    'bonds': {'ALL': {'NameOfCutoffFunction': 'cos',
                      'dcut': 0.02,
                      'radbase': 'SBessel',
                      'radparameters': [5.25],
                      'r_in': 0.5,
                      'delta_in': 0.1,
                      'core-repulsion': [10000, 0],
                      'rcut': 6.4},

              },
    "functions": {
            "number_of_functions_per_element": 1600,
            "ALL": {
                "nradmax_by_orders": [17,12,9,6,3],
                "lmax_by_orders": [0,9,7,3,1],
                "coeffs_init": ""
            },
    }
}

Si_bbasis=create_multispecies_basis_config(potential_config)
```

```python
Si_bbasis
Si_bbasis.total_number_of_functions
save_interim_potential(Si_bbasis, potential_filename='./empty_Si.yaml')
```
